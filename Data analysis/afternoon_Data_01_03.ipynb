```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd

import FiltradoMedia as fm #library for data processing
```

# Data display and adjustment

```python
a_1_03_82_halley = fm.datos_int(dir_data = '01_03_2020/afternoon/DATALOG_1800.TXT')
R_a_1_03_82_halley = fm.ajuste_int(a_1_03_82_halley)


fig = plt.figure(figsize=(18, 11))
plt.title('18:00', size = 20)
plt.xlabel('degrees [°]', fontsize = 20)                 
plt.ylabel('voltage [a.u]', fontsize = 20) 

plt.errorbar(a_1_03_82_halley[3],a_1_03_82_halley[1], yerr = a_1_03_82_halley[2], xerr = a_1_03_82_halley[4], linestyle=None, color='m', ms=5,fmt="s",capsize=5, capthick=1, label = r'$V^*$')
plt.plot(R_a_1_03_82_halley[0], R_a_1_03_82_halley[1], 'b-')
plt.grid()
plt.legend(fontsize = 18, loc=0)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.ylim((0,200))
plt.show()
```

```python
#norte-sur, calculation of the inclination of the polarization vector
pos_min = np.where(R_a_1_03_82_halley[1]==min(R_a_1_03_82_halley[1]))
R_a_1_03_82_halley[0][pos_min]
```
